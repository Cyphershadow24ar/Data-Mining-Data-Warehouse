```python
import pandas as pd
df = pd.read_csv('sales_dataset_custom_columns.csv', parse_dates=['Order Date'])
print(df)
display(df.head())
```

```python
slice_east =df[df['Region'] == 'East']
print(slice_east)
```

```python
dice_east_tv = df[(df['Region'] == 'East') & (df['Item'] == 'Television')]
print(dice_east_tv)
```

```python
rollup_region = df.groupby('Region').sum(numeric_only = True).reset_index()
print(rollup_region)
```

```python
rollup_manager = df.groupby('Manager').sum(numeric_only=True).reset_index()
print(rollup_manager)
```

```python
drilldown = df.groupby(['Region','Manager','Salesman'])['Units'].sum().reset_index()
print(drilldown)
```

```python
pivot = pd.pivot_table(df,
                       values='Units',
                       index=['Region', 'Manager'],
                       columns=['Item'],
                       aggfunc ='sum',
                       fill_value = 0)

print(pivot)
```
